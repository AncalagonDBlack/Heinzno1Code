```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from google.colab import files
import time
```

```python
#Tạo 1 file txt quăng hết url vô dạng cột
uploaded_file = files.upload()

# Lấy tên file
file_name = list(uploaded_file.keys())[0]

# Đọc nội dung file và chuyển thành danh sách url
with open(file_name, 'r') as file:
    url_list = [line.strip() for line in file.readlines()]

# In danh sách url
print(url_list)
```

```python
data = []

# Lặp qua từng URL để lấy dữ liệu
for url in url_list:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Tạo hàng cho URL, HTML và Value
    url_row = {'URL': url, 'HTML': 'URL', 'Value': url}
    data.append(url_row)

    # Lấy phần heading (h1-h5)
    headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5'])
    # delay cho đỡ bị chặn
    time.sleep(5)
    # Tạo hàng riêng cho từng heading
    for heading in headings:
        heading_text = heading.text.strip()
        heading_tag = heading.name.lower()  # Lấy tên thẻ heading (h1, h2, h3, h4, h5)
        row = {'URL': url, 'HTML': heading_tag, 'Value': heading_text}
        data.append(row)

    # Lấy phần paragraph (p)
    paragraphs = soup.find_all('p')

    # Tạo hàng riêng cho từng paragraph có giá trị
    for paragraph in paragraphs:
        paragraph_text = paragraph.text.strip()
        if paragraph_text:
            row = {'URL': url, 'HTML': 'Paragraph', 'Value': paragraph_text}
            data.append(row)

# Tạo DataFrame từ dữ liệu
df = pd.DataFrame(data)
```

```python
# Xuất dữ liệu ra file CSV
df.to_csv('output.csv', index=False)
files.download('output.csv')
```
